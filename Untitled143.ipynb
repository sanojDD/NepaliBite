````python
import tensorflow as tf
from tensorflow.keras import layers
import os

data_dir = 'nepali_food_data/dataset'
train_dir = os.path.join(data_dir, 'train')
test_dir = os.path.join(data_dir, 'test')

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomTranslation(0.1, 0.1),
    layers.RandomZoom(0.15),

    layers.RandomContrast(0.2),
    layers.RandomBrightness(0.1),

    layers.RandomTranslation(height_factor=0.1, width_factor=0.1, fill_mode='reflect'),
])

train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    label_mode='categorical',
    batch_size=32,
    image_size=(224, 224),
    shuffle=True
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    label_mode='categorical',
    batch_size=32,
    image_size=(224, 224),
    shuffle=False
)

class_names = train_ds.class_names

def prepare_dataset(ds, augment=False):
    # ds = ds.map(lambda x, y: (x / 255.0, y), num_parallel_calls=tf.data.AUTOTUNE)

    if augment:
        ds = ds.map(lambda x, y: (data_augmentation(x, training=True), y),
                    num_parallel_calls=tf.data.AUTOTUNE)

    return ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)

train_dataloader = prepare_dataset(train_ds, augment=True)
test_dataloader = prepare_dataset(test_ds, augment=False)

print(f"📊 Training classes found: {class_names}")
````

````python
from tensorflow.keras import regularizers

base_model = tf.keras.applications.MobileNetV3Large(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

inputs = tf.keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
x = tf.keras.applications.mobilenet_v3.preprocess_input(x)
x = base_model(x, training=False)

x = tf.keras.layers.GlobalAveragePooling2D()(x)

x = tf.keras.layers.BatchNormalization()(x)

x = tf.keras.layers.Dense(256, activation='relu',
                          kernel_regularizer=regularizers.l2(0.01))(x)

x = tf.keras.layers.Dropout(0.4)(x)

outputs = tf.keras.layers.Dense(len(class_names), activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)
````

````python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=7,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-6,
    verbose=1
)

checkpoint = ModelCheckpoint(
    'best_nepali_food_model.keras',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

callbacks_list = [early_stop, reduce_lr, checkpoint]
````

````python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_dataloader,
    validation_data=test_dataloader,
    epochs=50,
    callbacks=callbacks_list
)
````

````python
import tensorflow as tf
from tensorflow.keras import layers, regularizers, Input, Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import os

# --- 1. DATA LOADING ---
data_dir = 'nepali_food_data/dataset'
train_dir = os.path.join(data_dir, 'train')
test_dir = os.path.join(data_dir, 'test')

train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    label_mode='categorical',
    batch_size=32,
    image_size=(224, 224),
    shuffle=True
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    label_mode='categorical',
    batch_size=32,
    image_size=(224, 224),
    shuffle=False
)

class_names = train_ds.class_names

# Optimization: Removed augmentation from here to keep it on GPU via the Model
def prepare_dataset(ds):
    return ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE)

train_dataloader = prepare_dataset(train_ds)
test_dataloader = prepare_dataset(test_ds)

# --- 2. AUGMENTATION & MODEL ---
# Tuned for food: less extreme rotations, more focus on color/zoom
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomBrightness(0.1),
    layers.RandomContrast(0.1),
])

# Using EfficientNetV2B0 - Better accuracy than MobileNet
base_model = tf.keras.applications.EfficientNetV2B0(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

inputs = Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
# EfficientNetV2B0 has internal rescaling, so we don't need manual preprocessing!
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001))(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(len(class_names), activation='softmax')(x)

model = Model(inputs, outputs)

# --- 3. CALLBACKS ---
callbacks_list = [
    EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1),
    ModelCheckpoint('best_nepali_food_v2.keras', monitor='val_accuracy', save_best_only=True, verbose=1)
]

# --- 4. PHASE 1: INITIAL TRAINING ---
print("🚀 Starting Phase 1: Training the Head...")
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(
    train_dataloader,
    epochs=30,
    validation_data=test_dataloader,
    callbacks=callbacks_list
)

# --- 5. PHASE 2: FINE-TUNING ---
print("🚀 Starting Phase 2: Fine-Tuning the Backbone...")
base_model.trainable = True
# Freeze the bottom, unfreeze the top 20 layers
for layer in base_model.layers[:-20]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), # Tiny LR is a MUST
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    train_dataloader,
    epochs=20,
    validation_data=test_dataloader,
    callbacks=callbacks_list
)
````

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# --- 1. PLOT ACCURACY & LOSS GRAPHS ---
def plot_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 5))

    # Accuracy Plot
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.legend(loc='lower right')

    # Loss Plot
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.legend(loc='upper right')
    plt.show()

# --- 2. GENERATE CONFUSION MATRIX & TABLE ---
def show_confusion_matrix(model, test_ds, class_names):
    # Get all true labels and predictions
    y_true = []
    y_pred = []

    for images, labels in test_ds:
        preds = model.predict(images, verbose=0)
        y_true.extend(np.argmax(labels, axis=1))
        y_pred.extend(np.argmax(preds, axis=1))

    # Print Classification Table (Precision, Recall, F1)
    print("\n📊 CLASSIFICATION REPORT")
    print(classification_report(y_true, y_pred, target_names=class_names))

    # Plot Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix: Nepali Food Classifier')
    plt.ylabel('Actual Food')
    plt.xlabel('Predicted Food')
    plt.show()

# Call the functions
# (Replace 'history' with your history object from fit())
# plot_history(history)
show_confusion_matrix(model, test_dataloader, class_names)
````

````python
from sklearn.metrics import precision_recall_curve, roc_curve, auc
from sklearn.preprocessing import label_binarize

def plot_multiclass_roc(model, test_ds, class_names):
    # 1. Get predictions and true labels
    y_true_list = []
    y_pred_list = []
    for images, labels in test_ds:
        y_true_list.extend(np.argmax(labels, axis=1))
        y_pred_list.extend(model.predict(images, verbose=0))

    y_true = label_binarize(y_true_list, classes=range(len(class_names)))
    y_pred = np.array(y_pred_list)

    # 2. Plotting
    plt.figure(figsize=(10, 7))
    for i in range(len(class_names)):
        fpr, tpr, _ = roc_curve(y_true[:, i], y_pred[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{class_names[i]} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve for Nepali Food Classes')
    plt.legend(loc='lower right')
    plt.show()

# Call the function
plot_multiclass_roc(model, test_dataloader, class_names)
````

## **Nepali Food Image Classifier** 🍛🚀

This project is a high-performance deep learning system designed to identify various Nepali dishes from images. It specifically addresses the visual diversity of Nepali cuisine, from the unique circular shape of **Selroti** to the complex arrangements of **Dalbhat**.

---

### **Project Overview**

* **Objective:** To build an accurate classification model for 9 distinct types of popular Nepali food.
* **Core Model:** **EfficientNetV2B0** (Transfer Learning).
* **Performance:** Achieved **90% Validation Accuracy** with an average **F1-Score of 0.87**.
* **Developer:** Sanoj.

---

### **Dataset Classes**

The model is trained to recognize the following 9 classes:

1. **Chatamari** (Newari "pizza")
2. **Chhoila** (Spiced grilled meat)
3. **Dalbhat** (The staple lentil-rice set)
4. **Dhindo** (Traditional buckwheat/millet porridge)
5. **Gundruk** (Fermented leafy green)
6. **Kheer** (Rice pudding)
7. **Momo** (Dumplings)
8. **Sekuwa** (Roasted meat)
9. **Selroti** (Ring-shaped sweet bread)

---

### **Model Architecture & Training**

The training was conducted in two distinct phases to maximize feature extraction while preventing overfitting:

* **Phase 1: Head Training:** The base **EfficientNetV2B0** layers were frozen, and a custom global average pooling and dense head were trained. Accuracy reached **92.4%** at the peak of this phase.
* **Phase 2: Fine-Tuning:** The top layers of the backbone were unfrozen with a very low learning rate ($1.0 \times 10^{-5}$) to specialize the model on specific Nepali food textures.
* **Hardware Used:** 11th Gen Intel i5-1135G7, 8GB RAM (Development Environment: Google Colab).

---

### **Key Metrics (Final Result)**

Based on the **Classification Report**, the model shows exceptional reliability for iconic dishes:

| Food Item | F1-Score | Note |

| **Selroti** | **1.00** | Perfect detection of shape. |

````python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

# 1. Load the "Winner" Model
model_path = 'best_nepali_food_v2.keras'
model = tf.keras.models.load_model(model_path)

# 2. Define the Class Names (Must match your training order!)
class_names = [
    'chatamari', 'chhoila', 'dalbhat', 'dhindo',
    'gundruk', 'kheer', 'momo', 'sekuwa', 'selroti'
]

def predict_nepali_food(img_path):
    # Load and preprocess the image (EfficientNetV2B0 expects 224x224)
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Create a batch

    # Note: EfficientNetV2B0 has rescaling built-in, so we don't need /255.0 manually

    # Make Prediction
    predictions = model.predict(img_array)
    score = predictions[0] # No need for tf.nn.softmax here # Convert to probabilities

    # Get results
    predicted_class = class_names[np.argmax(score)]
    confidence = 100 * np.max(score)

    # Display the result
    plt.imshow(img)
    plt.title(f"Prediction: {predicted_class} ({confidence:.2f}%)")
    plt.axis('off')
    plt.show()

    print(f"This image most likely belongs to {predicted_class} with a {confidence:.2f} percent confidence.")


predict_nepali_food('/content/momo.png')
````
